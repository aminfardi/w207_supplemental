```python
# create a Python list of feature names

feature_cols = ['Main_Mass_Flow', 'Additive_1_Ratio', 'Additive_2_Ratio', 'Additive_3_Ratio','Additive_4_Ratio', 
                'Additive_5_Ratio', 'Additive_6_Ratio','Flow_Gas_Ratio','Pressure', 'Blending']

X = SAP_data[feature_cols]

# print the first 5 rows
X.head()
```

```python
# check the type and shape of X
print(type(X))
print(X.shape)
```

```python

# select a Series from the DataFrame
y = SAP_data['Quality']

# print the first 5 values
y.head()
```

```python
# check the type and shape of y
print(type(y))
print(y.shape)
```

## Splitting X and y into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# default split is 75% for training and 25% for testing
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## Data Standarization

```python
# standarization
print(X_train.shape, y_train.shape)
X_train_mean = np.mean(X_train, axis=0)
X_train_std = np.std(X_train, axis=0)


# dataframe doesn't need reshape
# numpy array needs reshape. 
# reshape(1,-1)
X_train_S = (X_train - X_train_mean) / X_train_std

X_test_S = (X_test - X_train_mean) / X_train_std

print(X_train_S.shape, X_test_S.shape)
```

```python
X_train_S.head()
```

Observetion: 

> "Pressure" becomes "NaN" during standarization. I should remove it from data set since it contributes nothing to final result.

```python
del X_train_S['Pressure']
del X_test_S['Pressure']
X_train_S.head()
```

## Anomaly Detection

Assume a set of data {$X_{0}$, $X_{1}$, ..., $X_{n}$}

for loop each element of data:

    assume $X_{k}$ is current selection
    
    calculate mean and std of all others.
    
    if $X_{k}$ -mean> 3 * std, then it's a outlier


## Linear regression

```python
# standarization
# X_train, X_test
# Y_train, Y_test
# tip #1, don't standarize entire dataset, only standarize train data set.
# tip #2, don't standarize train and test data seperately.
# tip #3, standarize train data and keep mean/std for test data standarization.
print(X_train.shape, y_train.shape)
X_train_mean = np.mean(X_train, axis=0)
X_train_std = np.std(X_train, axis=0)


# dataframe doesn't need reshape
# numpy array needs reshape. 
# reshape(1,-1)
X_train_S = (X_train - X_train_mean) / X_train_std

X_test_S = (X_test - X_train_mean) / X_train_std

print(X_train_S.shape, X_test_S.shape)
```

```python
# instantiate
linreg = LinearRegression()

# fit the model to the training data (learn the coefficients)
linreg.fit(X_train_S, y_train)
```

```python
# print the intercept and coefficients
print(linreg.intercept_)
print(linreg.coef_)
```

```python
# pair the feature names with the coefficients
list(zip(feature_cols, linreg.coef_))
```

```python
# make predictions on the testing set
y_pred = linreg.predict(X_test_S)
```

```python
#RMSE
print(y_pred[:10])
print(y_test[:10])
```
